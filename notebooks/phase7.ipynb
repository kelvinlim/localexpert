```
from statsmodels.stats.power import TTestIndPower
import math

# A-priori power analysis for two-group independent t-test
# Parameters:
# - effect_size (Cohen's d) = 0.5
# - alpha = 0.05 (two-sided)
# - target power = 0.80
# - equal allocation (ratio = 1)

power_analysis = TTestIndPower()

# Solve for sample size (nobs1 = per-group sample size)
result = power_analysis.solve_power(
    effect_size=0.5,
    nobs=None,      # Leave None to solve for sample size
    alpha=0.05,
    power=0.80,
    ratio=1         # Equal allocation between groups
)

print(f"Required sample size per group (nobs1): {result}")
print(f"Total sample size: {int(math.ceil(result)) * 2}")
print(f"\nRounded up per group: {math.ceil(result)}")
print(f"Rounded up total N: {math.ceil(result) * 2}")

# Also compute achieved power at this rounded-up N to confirm
rounded_n = math.ceil(result)
result_check = power_analysis.solve_power(
    effect_size=0.5,
    nobs=rounded_n,
    alpha=0.05,
    power=None,
    ratio=1
)
print(f"\nAchieved power at rounded-up N: {result_check:.4f}")
```
